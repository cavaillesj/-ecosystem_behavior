# Contribution of a times series in the average variability

```python
Number_of_simulation = 20
ind = np.arange(Number_of_simulation)
NumberPoint = 10

Freq = [0.001, 0.05, 0.1, 0.2, 0.3, 0.4, 0.6, 2]
Var = np.zeros((len(Freq), NumberPoint, Number_of_simulation))

for j, freq in enumerate(Freq):    
    Param_freq = {"p":  freq}
    Fire_param["param_freq"] = Param_freq
    for i in range(NumberPoint):
        for k in range(Number_of_simulation):
            var = np.NaN
            c = 0
            while(np.isnan(var) and c < 10):
                c+=1
                O = Ode(finalTime = finalTime, Fire_param = Fire_param, dt = dt, Param_phy=Param_phy, Init=Init)
                O.solve_by_part()
                var = O.variability()
            if(c==10):
                print("Point : "+str(i)+", simulation : "+str(k)+", Always nan")
            Var[j,i,k] = var
        Var[j,i] = np.sort(Var[j,i])
```

```python
plt.figure(figsize = (15, 30))
mmax = np.max(np.max(np.nan_to_num(Var)))
for j, freq in enumerate(Freq):
    plt.subplot(4,2,j+1)
    for i in range(NumberPoint):
        plt.plot(ind, Var[j, i], "+", color = "blue")
        plt.ylim(0, mmax)
    plt.bar(ind, np.nanmean(Var[j], axis = 0))
    plt.ylim(0, mmax)
    plt.title("Distribution of the variability\nfrequency = "+str(freq))
    plt.ylabel("Variability")
    plt.xlabel("Simulation")
    plt.xticks(ind, ind+1)
plt.ylim(0, mmax)
plt.show()
```

### Boxplot

```python
Number_of_simulation = 20
ind = np.arange(Number_of_simulation)
NumberPoint = 10

Freq = [0.001, 0.05, 0.1, 0.2, 0.3, 0.4, 0.6, 2]
VarBoxplot = np.zeros((len(Freq), NumberPoint, Number_of_simulation))
ColBoxplot = np.zeros_like(VarBoxplot)

for j, freq in enumerate(Freq):    
    Param_freq = {"p":  freq}
    Fire_param["param_freq"] = Param_freq
    for i in range(NumberPoint):
        S = Solver(numberOfSimulation = Number_of_simulation, finalTime = finalTime, Fire_param = Fire_param, dt = dt, Param_phy=Param_phy, Init=Init)
        S.run()
        for k in range(Number_of_simulation):
            VarBoxplot[j,i,k] = S.O[k].variability()
            ColBoxplot[j,i,k] = S.O[k].get_collapse()
     #   arg_sort = np.argsort(VarBoxplot[j,i])
      #  VarBoxplot[j,i] = VarBoxplot[j,i][arg_sort]
       # ColBoxplot[j,i] = ColBoxplot[j,i][arg_sort]
```

```python
Freq_label = []
for freq in Freq:
    Freq_label += [str(freq)]
```

```python
M = np.nanmean(VarBoxplot, axis = 1)
M2 = []
for i in range(len(M)):
    M2 += [[]]
    for j in range(len(M[0])):
        if(1-np.isnan(M[i,j])):
            M2[i] += [M[i,j]] 
    
#M = np.nanmean(Var, axis = 1).transpose()
```

```python
plt.figure(figsize = (16, 8))
plt.boxplot(M2)#, labels=Freq_label[:-2:])
plt.xlabel("Frequency")
plt.ylabel("Variability")
plt.title("Distribution of the variability")
plt.show()
```

### Without biaisis

```python
Var_without_biaisis = np.zeros((len(Freq), NumberPoint))
for j, freq in enumerate(Freq):    
    Param_freq = {"p":  freq}
    Fire_param["param_freq"] = Param_freq
    for i in range(NumberPoint):
        var_after = np.mean(VarBoxplot[j,i][np.isnan(ColBoxplot[j,i])])
        var_no_collapse = np.mean(VarBoxplot[j,i][1-np.isnan(ColBoxplot[j,i])])
        Var_without_biaisis[j,i] = (var_after*sum(np.isnan(ColBoxplot[j,i])) + var_no_collapse*sum(1-np.isnan(ColBoxplot[j,i]))) / len(VarBoxplot[j,i])
```

```python
M = Var_without_biaisis
M2 = []
for i in range(len(M)):
    M2 += [[]]
    for j in range(len(M[0])):
        if(1-np.isnan(M[i,j])):
            M2[i] += [M[i,j]] 
    
#M = np.nanmean(Var, axis = 1).transpose()
```
